```python
import json
import csv
```

```python
f = open('water_portability.json')
lakesData = json.load(f)
```

```python
matches = 0

with open('lakes-volume.csv', newline='', encoding='utf-8-sig') as csvfile:
    reader = csv.DictReader(csvfile)
    # lakes = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in reader:
        # print(row)
        print('currently in ', row['Name'])
        for lake in lakesData:
            # print(lakesData[lake]['name'])
            if row['Name'] == lakesData[lake]['name'] or lakesData[lake]['name'] in row['Name']:
                if lakesData[lake]["portability(m3)"] == 'undefined':
                    lakesData[lake]["portability(m3)"] = int(row['Volume'])
                    matches = matches + 1
                    print('match!')

print(matches)
```
